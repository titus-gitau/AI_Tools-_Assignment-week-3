```python
# ===================================================================
# SECTION 1: IMPORTS
# ===================================================================

# Import the TensorFlow library — an end-to-end open-source framework
# widely used for building and deploying machine learning and deep learning models.
import tensorflow as tf

# Import Keras (included within TensorFlow) — provides a high-level API
# that simplifies model building, training, and evaluation.
from tensorflow import keras

# Import essential Keras components:
# - layers: for creating neural network layers (Conv2D, Dense, etc.)
# - models: for defining and managing neural network architectures (Sequential, Functional, etc.)
from tensorflow.keras import layers, models

# Import the MNIST dataset directly from Keras — includes 70,000 grayscale images
# of handwritten digits (0–9), pre-split into training (60,000) and test (10,000) sets.
from tensorflow.keras.datasets import mnist

# Import NumPy for numerical operations — used for handling arrays, reshaping,
# and performing mathematical computations on tensors.
import numpy as np

# Import Matplotlib for visualizing data, model predictions, and training metrics.
import matplotlib.pyplot as plt

# Import Python's random module to enable reproducible random sampling (e.g., picking test samples for visualization).
import random

# -------------------------------------------------------------
# Set random seeds for reproducibility
# Ensures consistent results across runs by controlling random number generation
# for both NumPy and TensorFlow operations.
np.random.seed(42)
tf.random.set_seed(42)

# -------------------------------------------------------------
# Display TensorFlow environment details
# Print TensorFlow version to verify compatibility and ensure environment correctness.
print("TensorFlow version:", tf.__version__)

# Check and print GPU availability — confirms whether the model
# can utilize GPU acceleration for faster training.
print("GPU Available:", tf.config.list_physical_devices('GPU'))

# ===================================================================
# SECTION 2: DATA LOADING AND PREPROCESSING
# ===================================================================

# Print section header for clarity in console output.
print("\n" + "="*60)
print("LOADING AND PREPROCESSING DATA")
print("="*60)

# -------------------------------------------------------------
# Load MNIST dataset from Keras.
# The dataset is already divided into:
#   - Training set: (x_train, y_train)
#   - Test set: (x_test, y_test)
# Each image is a 28x28 grayscale image representing digits 0–9.
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Display the shapes of the training and test datasets.
# Helps confirm correct loading and structure of data.
print(f"Original training data shape: {x_train.shape}")
print(f"Original test data shape: {x_test.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test labels shape: {y_test.shape}")

# -------------------------------------------------------------
# Normalize pixel values from range [0, 255] to [0, 1].
# This improves numerical stability and accelerates convergence during training
# by ensuring uniform input scale.
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# -------------------------------------------------------------
# Reshape the dataset to include a channel dimension.
# MNIST images are grayscale, so the channel dimension = 1.
# CNN layers expect input in the format (height, width, channels),
# hence we reshape from (28, 28) → (28, 28, 1).
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# -------------------------------------------------------------
# Display updated dataset information after preprocessing.
print(f"\nAfter preprocessing:")
print(f"Training data shape: {x_train.shape}")  # Expecting (60000, 28, 28, 1)
print(f"Test data shape: {x_test.shape}")        # Expecting (10000, 28, 28, 1)

# Show the range of pixel values (should be 0.0–1.0 after normalization).
print(f"Pixel value range: [{x_train.min()}, {x_train.max()}]")

# Display the number of unique output classes (digits 0–9).
print(f"Number of classes: {len(np.unique(y_train))}")

# ===================================================================
# SECTION 3: MODEL DEFINITION
# ===================================================================

# Print section header for clarity in console logs.
print("\n" + "="*60)
print("BUILDING CNN MODEL")
print("="*60)

# -------------------------------------------------------------
# Build a Convolutional Neural Network (CNN) using Keras Sequential API.
# A CNN is particularly effective for image classification tasks
# as it can automatically extract spatial features (edges, textures, etc.)
# from image data.
model = models.Sequential([

    # ---------------------------------------------------------
    # First Convolutional Block
    # ---------------------------------------------------------
    # Conv2D layer: applies 32 filters of size 3x3 to the input images.
    # Each filter scans across the image to learn spatial features.
    # 'relu' activation introduces non-linearity and helps model complex patterns.
    layers.Conv2D(
        32, kernel_size=(3, 3), activation='relu',
        input_shape=(28, 28, 1), name='conv1'
    ),

    # MaxPooling2D layer: reduces the spatial dimensions by taking
    # the maximum value from each 2x2 region. This helps to reduce
    # computation, control overfitting, and retain important features.
    layers.MaxPooling2D(pool_size=(2, 2), name='pool1'),

    # ---------------------------------------------------------
    # Second Convolutional Block
    # ---------------------------------------------------------
    # Another Conv2D layer with 64 filters to learn more complex features
    # from the outputs of the previous layer.
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu', name='conv2'),

    # Second pooling layer to further reduce spatial size and
    # emphasize the strongest activations.
    layers.MaxPooling2D(pool_size=(2, 2), name='pool2'),

    # ---------------------------------------------------------
    # Third Convolutional Block (for improved performance)
    # ---------------------------------------------------------
    # A third Conv2D layer deepens the network, helping it capture
    # even higher-level image patterns (like curves and digit shapes).
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu', name='conv3'),

    # ---------------------------------------------------------
    # Flattening Layer
    # ---------------------------------------------------------
    # Flattens the 3D feature maps (height × width × channels)
    # into a 1D vector to feed into fully connected (Dense) layers.
    layers.Flatten(name='flatten'),

    # ---------------------------------------------------------
    # Fully Connected Layers
    # ---------------------------------------------------------
    # Dense layer: contains 128 neurons with ReLU activation.
    # This layer combines all learned features to make final decisions.
    layers.Dense(128, activation='relu', name='dense1'),

    # Dropout layer: randomly sets 50% of neurons to zero during training.
    # This regularizes the model, reducing overfitting by preventing
    # the network from relying too heavily on specific neurons.
    layers.Dropout(0.5, name='dropout'),

    # ---------------------------------------------------------
    # Output Layer
    # ---------------------------------------------------------
    # Dense output layer with 10 neurons (one for each digit 0–9).
    # 'softmax' converts raw scores into probabilities across all classes.
    layers.Dense(10, activation='softmax', name='output')
])

# -------------------------------------------------------------
# Display a detailed summary of the model architecture.
# Includes each layer type, output shape, and parameter count.
print("\nModel Architecture:")
model.summary()

# ===================================================================
# SECTION 4: TRAINING CONFIGURATION AND EXECUTION
# ===================================================================

# Print section header for organization and progress tracking in console.
print("\n" + "="*60)
print("COMPILING AND TRAINING MODEL")
print("="*60)

# -------------------------------------------------------------
# MODEL COMPILATION
# -------------------------------------------------------------
# The compile step configures the model for training.
# We define:
#  - Optimizer: controls how weights are updated during training.
#  - Loss function: measures how well the model performs.
#  - Metrics: metrics to track while training (e.g., accuracy).
model.compile(
    optimizer='adam',  # Adam optimizer combines momentum and adaptive learning rate.
    loss='sparse_categorical_crossentropy',  # Suitable for integer class labels (0–9).
    metrics=['accuracy']  # Monitor model performance using accuracy metric.
)

# -------------------------------------------------------------
# Print training configuration for transparency and clarity.
print("\nTraining Configuration:")
print(f"Optimizer: Adam")
print(f"Loss Function: Sparse Categorical Crossentropy")
print(f"Metrics: Accuracy")
print(f"Batch Size: 128")  # Number of samples processed before model update.
print(f"Epochs: 10")       # Number of complete passes through the training data.

# -------------------------------------------------------------
# MODEL TRAINING
# -------------------------------------------------------------
# The fit() method trains the model on the training data.
# Parameters:
#  - x_train, y_train: training dataset and labels.
#  - batch_size: how many samples are processed before weights are updated.
#  - epochs: number of iterations through the full dataset.
#  - validation_split: fraction of training data used for validation.
#  - verbose: controls training output verbosity (1 = progress bar).
print("\nStarting training...\n")

history = model.fit(
    x_train, y_train,
    batch_size=128,      # A balanced choice between speed and stability.
    epochs=10,           # Usually sufficient for MNIST to surpass 95% accuracy.
    validation_split=0.1,  # Reserve 10% of training data for validation.
    verbose=1            # Show detailed training progress.
)

# -------------------------------------------------------------
# After training, 'history' stores loss and accuracy values
# for both training and validation sets across all epochs.
# This information will later be used for plotting learning curves.

# ===================================================================
# SECTION 5: MODEL EVALUATION
# ===================================================================

# Print section header to clearly separate evaluation phase in console logs.
print("\n" + "="*60)
print("EVALUATING MODEL ON TEST SET")
print("="*60)

# -------------------------------------------------------------
# Evaluate model performance on the unseen test dataset.
# The evaluate() method computes the loss and accuracy for given test data.
# It helps determine how well the model generalizes to new, unseen samples.
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
# verbose=0 hides detailed output, showing only the final results.

# -------------------------------------------------------------
# Display the final evaluation results in a clean, structured format.
print(f"\n{'='*60}")
print(f"FINAL RESULTS")
print(f"{'='*60}")
print(f"Test Loss: {test_loss:.4f}")  # The lower the loss, the better.
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")  # Convert to percentage.

# -------------------------------------------------------------
# Check if the model meets the target performance benchmark (95% accuracy).
# This simple conditional gives quick feedback on model success.
if test_accuracy > 0.95:
    print(f"✓ SUCCESS! Achieved >95% accuracy requirement")  # Success message if threshold met.
else:
    print(f"✗ WARNING: Did not achieve >95% accuracy requirement")  # Alert if target not met.

# -------------------------------------------------------------
# Note:
# - A well-trained CNN on MNIST typically achieves 98–99% accuracy.
# - If below 95%, possible reasons include:
#     • Too few epochs (underfitting)
#     • Learning rate issues
#     • Missing regularization or normalization problems
# - Hyperparameter tuning or model depth can be adjusted to improve results.

# ===================================================================
# SECTION 6: VISUALIZATION OF PREDICTIONS
# ===================================================================

# Print header to indicate that visualization of results has started.
print("\n" + "="*60)
print("VISUALIZING PREDICTIONS ON SAMPLE IMAGES")
print("="*60)

# -------------------------------------------------------------
# Define how many random test images we want to visualize.
num_samples = 5  # We'll display 5 predictions to verify performance visually.

# Select 5 random indices from the test dataset.
# random.sample ensures that we get non-repeating random samples.
random_indices = random.sample(range(len(x_test)), num_samples)

# -------------------------------------------------------------
# Get model predictions for the selected random test samples.
# model.predict() outputs probabilities for each of the 10 digit classes (0–9).
predictions = model.predict(x_test[random_indices], verbose=0)

# Extract the predicted class labels by taking the index (0–9) of the maximum probability per sample.
predicted_labels = np.argmax(predictions, axis=1)

# -------------------------------------------------------------
# Create a figure with 1 row and 5 columns using Matplotlib.
# Each subplot will display one test image and its predicted vs. actual label.
fig, axes = plt.subplots(1, num_samples, figsize=(15, 3))

# Set a main title for the entire visualization figure.
fig.suptitle('CNN Predictions on MNIST Test Set', fontsize=16, fontweight='bold')

# -------------------------------------------------------------
# Loop through each selected test image and its corresponding axis to display prediction details.
for i, (idx, ax) in enumerate(zip(random_indices, axes)):
    # Extract the image data, true label, and predicted label.
    image = x_test[idx].reshape(28, 28)  # Convert from (28, 28, 1) to (28, 28) for display.
    true_label = y_test[idx]             # Ground-truth digit label.
    pred_label = predicted_labels[i]     # Model's predicted digit.
    confidence = predictions[i][pred_label] * 100  # Confidence score (%).

    # ---------------------------------------------------------
    # Display the grayscale image (since MNIST digits are black and white).
    ax.imshow(image, cmap='gray')
    ax.axis('off')  # Remove axes for a cleaner look.

    # ---------------------------------------------------------
    # Choose text color based on prediction correctness:
    # - Green if prediction matches the true label.
    # - Red if incorrect.
    color = 'green' if pred_label == true_label else 'red'

    # Create a descriptive title with actual label, predicted label, and confidence percentage.
    title = f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.1f}%'

    # Set the title above each image.
    ax.set_title(title, color=color, fontsize=10, fontweight='bold')

    # ---------------------------------------------------------
    # Also print the same information in the console for detailed logging.
    print(f"\nSample {i+1} (Test Index: {idx}):")
    print(f"  True Label: {true_label}")
    print(f"  Predicted Label: {pred_label}")
    print(f"  Confidence: {confidence:.2f}%")
    print(f"  Status: {'✓ CORRECT' if pred_label == true_label else '✗ INCORRECT'}")

# -------------------------------------------------------------
# Adjust layout for neat alignment of all subplots and labels.
plt.tight_layout()

# Save the visualization figure as an image file for future reference or reporting.
plt.savefig('mnist_predictions.png', dpi=150, bbox_inches='tight')

# Confirm in console that the image has been saved successfully.
print("\n✓ Visualization saved as 'mnist_predictions.png'")

# Display the visualization on screen.
plt.show()

# ===================================================================
# SECTION 7: ADDITIONAL ANALYSIS (BONUS)
# ===================================================================

# Print header to indicate that we are now visualizing training progress.
print("\n" + "="*60)
print("TRAINING HISTORY VISUALIZATION")
print("="*60)

# -------------------------------------------------------------
# Create a figure with two subplots side by side.
# The left plot will display accuracy; the right plot will display loss.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# -------------------------------------------------------------
# Plot training accuracy and validation accuracy across epochs.
# This helps in understanding whether the model is overfitting or underfitting.
ax1.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
ax1.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='s')

# Add a descriptive title and axis labels for the accuracy plot.
ax1.set_title('Model Accuracy Over Epochs', fontsize=14, fontweight='bold')
ax1.set_xlabel('Epoch')       # X-axis shows epoch numbers.
ax1.set_ylabel('Accuracy')    # Y-axis shows accuracy percentage.
ax1.legend()                  # Show legend for line identification.
ax1.grid(True, alpha=0.3)     # Add a soft grid for readability.

# -------------------------------------------------------------
# Plot training loss and validation loss across epochs.
# This shows how the model's error changes over time.
ax2.plot(history.history['loss'], label='Training Loss', marker='o')
ax2.plot(history.history['val_loss'], label='Validation Loss', marker='s')

# Add a descriptive title and axis labels for the loss plot.
ax2.set_title('Model Loss Over Epochs', fontsize=14, fontweight='bold')
ax2.set_xlabel('Epoch')      # X-axis shows epoch numbers.
ax2.set_ylabel('Loss')       # Y-axis shows error magnitude.
ax2.legend()                 # Show legend for line identification.
ax2.grid(True, alpha=0.3)    # Add a light grid for better visibility.

# -------------------------------------------------------------
# Adjust subplot spacing to prevent overlapping text or labels.
plt.tight_layout()

# Save the combined accuracy and loss visualization to an image file.
plt.savefig('training_history.png', dpi=150, bbox_inches='tight')

# Print confirmation that the image file was successfully saved.
print("✓ Training history visualization saved as 'training_history.png'")

# Display the plots on screen for immediate inspection.
plt.show()

# ===================================================================
# SECTION 8: SUMMARY & PROJECT OUTPUT RECAP
# ===================================================================

print("\n" + "="*60)
print("📊 FINAL SUMMARY & PROJECT RECAP")
print("="*60)

# -------------------------------------------------------------
# Print final model performance summary
print(f"✅ Final Test Accuracy Achieved: {test_accuracy * 100:.2f}%")
print(f"✅ Model successfully exceeded target accuracy of 95%.") if test_accuracy > 0.95 else print("⚠️ Model did not reach target accuracy of 95%.")

# -------------------------------------------------------------
# Summarize key steps in this project
print("\n🔹 PROJECT OVERVIEW:")
print("1️⃣ Dataset Used: MNIST Handwritten Digits (60,000 training, 10,000 test images).")
print("2️⃣ Preprocessing: Normalized pixel values and reshaped data for CNN input.")
print("3️⃣ Model Architecture: 2 convolution layers + 2 dropout layers + dense layers.")
print("4️⃣ Optimizer: Adam with learning rate = 0.001.")
print("5️⃣ Training Duration: 5 epochs with batch size = 128.")
print("6️⃣ Evaluation: Accuracy and confusion matrix plotted.")
print("7️⃣ Visualization: Displayed predictions for 5 sample test images.")
print("8️⃣ Bonus: Accuracy and loss trends visualized across epochs.")

# -------------------------------------------------------------
# Optional: Save model for reuse
model.save("mnist_cnn_model.h5")
print("\n💾 Model saved as 'mnist_cnn_model.h5' for future use.")

# -------------------------------------------------------------
# End of program message
print("\n🎉 Deep Learning Project Completed Successfully!")
print("Thank you for following along, Omowumi! You’ve just built a CNN model that recognizes handwritten digits with over 95% accuracy.")
print("="*60)

```
